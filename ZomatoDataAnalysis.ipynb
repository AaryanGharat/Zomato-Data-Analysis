```python
import os
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
pwd = os.getcwd()
filepath=os.path.join(pwd,"dataset/zomato.csv").replace("\\", "/")
```

```python
df=pd.read_csv(filepath)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.describe()
```

```python
df.columns
```

```python
df=df.drop(['url', 'address','phone', 'menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=="NEW" or value=="-"):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df['Cost2plates'].unique()
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2plates']=df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].head()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type=df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_type_lessthan1000=rest_type[rest_type<1000]
rest_type_lessthan1000
```

```python
def handleresttype(value):
    if(value in rest_type_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(handleresttype)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
location_lessthan300=location[location<300]


def handlelocation(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df['location']=df['location'].apply(handlelocation)
df['location'].value_counts()
```

```python
df.head()
```

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisines_lessthan300=cuisines[cuisines<300]


def handlecuisines(value):
    if(value in cuisines_lessthan300):
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(handlecuisines)
df['cuisines'].value_counts()
```
